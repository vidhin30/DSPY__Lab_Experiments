## **Notebook 2:** Exploratory Data Analysis (EDA) – Iris Dataset

**1. Introduction**:
This notebook performs:

* Observation
* Conclusion

**2. Import Libraries**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**3. Load Cleaned Dataset**

```python
import seaborn as sns

# Load dataset
df = sns.load_dataset('iris')

# Rename columns to match your project
df.rename(columns={
    'sepal_length': 'SepalLength',
    'sepal_width': 'SepalWidth',
    'petal_length': 'PetalLength',
    'petal_width': 'PetalWidth',
    'species': 'Species'
}, inplace=True)

# Save file
df.to_csv('iris_cleaned.csv', index=False)

print("File saved as iris_cleaned.csv")
```

```python
df.info()
df.describe()
```

## **NON-GRAPHICAL EDA**
**Analysis 1:** Summary

**Objective:**

To understand central tendency and spread of numerical features

**Type:**

Univariate Non-Graphical Analysis

```python
df.describe()
```

**Observation:**
SepalLength ranges roughly from 4.3 to 7.9
PetalLength shows wider variation
PetalWidth varies significantly across species

**Conclusion:**
Petal measurements have more variability and may be useful for classification

**Analysis 2:** Group-wise Mean (Species-wise Analysis)

**Objective:**
To compare average feature values across species

**Type:**
Multivariate Non-Graphical Analysis

```python
df.groupby('Species').mean()
```

**Observation:**
Setosa has smallest petal length and width
Virginica has largest values
Versicolor lies in between

**Conclusion:**
Petal features strongly differentiate species

# **GRAPHICAL EDA**
**Analysis 3:** Distribution Plot

**Objective:**
To understand distribution of Sepal Length

**Type:**
Univariate Graphical Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(df['SepalLength'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

**Observation:**
Data is approximately normally distributed
Slight skewness present

**Conclusion:**
Sepal length alone may not clearly separate species

**Analysis 4:** Box Plot (Species vs Petal Length)

**Objective:**
To detect variation and outliers across species

**Type:**
Multivariate Graphical Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Species', y='PetalLength', data=df)
plt.title('Petal Length across Species')
plt.show()
```

**Observation:**
Clear separation between species
Setosa has very small petal length
Virginica has highest
**Conclusion:**
Petal Length is a strong discriminative feature

**Analysis 5:** Scatter Plot

**Objective:**
To examine relationship between petal length and width

**Type:**
Multivariate Graphical Analysis

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='PetalLength', y='PetalWidth', hue='Species', data=df)
plt.title('Petal Length vs Petal Width')
plt.show()
```

**Observation:**

* Strong positive correlation
* Clear clustering of species

**Conclusion:**
These features are highly useful for classification

**Analysis 6:** Correlation Matrix

**Objective:**
To measure relationships between numerical variables

**Type:**
Multivariate Graphical Analysis

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop('Species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**Observation:**

* PetalLength and PetalWidth are highly correlated
* Sepal features show weaker correlation

**Conclusion:**
Petal features are more informative

# **Conclusion**
Petal features (length & width) are most important
Sepal features are less discriminative
Dataset shows clear clustering, making it ideal for classification tasks
