**Q1. What is 2 tailed and  tailed T Test?**

A T-test is a statistical hypothesis test that is used to determine if there is a significant difference between the means of two groups. It is often used when the sample size is small (typically less than 30) or when the population standard deviation is unknown.

**Tailed T-Test (One-Sided):**

A 1-tailed T-test is used when you have a specific direction in mind for your hypothesis. You are only interested in whether the sample mean is significantly greater than OR significantly less than the hypothesized population mean, but not both. For example:
* Scenario: You want to test if a new
teaching method increases students' test scores.
* Hypotheses:
  1. Null Hypothesis (H0): The new teaching method has no effect on test scores (or decreases them).
  2. Alternative Hypothesis (Ha): The new teaching method increases test scores.
* Critical Region: The critical region for rejecting the null hypothesis is entirely in one tail of the t-distribution (either the right tail for an increase or the left tail for a decrease).



2-Tailed T-Test (Two-Sided):

A 2-tailed T-test is used when you want to test for any significant difference between the sample mean and the hypothesized population mean, regardless of the direction. You are interested in whether the sample mean is significantly different (either greater or less) from the population mean. For example:

 * Scenario: You want to test if there
is any difference in test scores between two different groups of students.
* Hypotheses:
    1.  Null Hypothesis (H0): There is no difference in test scores between the two groups.
    2.  Alternative Hypothesis (Ha): There is a significant difference in test scores between the two groups.
* Critical Region: The critical region for rejecting the null hypothesis is split between both tails of the t-distribution.

**Q2. Explain what is Type I and Type II errors with an example.**

Type I and Type II errors in hypothesis testing :
In hypothesis testing, we formulate two competing hypotheses:

  1. Null Hypothesis (H0): A statement of no effect or no difference. We assume this is true until we have enough evidence to reject it.
  2. Alternative Hypothesis (Ha): A statement that contradicts the null hypothesis, suggesting an effect or difference.

Our goal is to decide whether to reject H0 or fail to reject H0 based on sample data.

**Type I Error (False Positive)**

* Definition: A Type I error occurs
when you reject a true Null Hypothesis. In simpler terms, you conclude there is a significant effect or difference when, in reality, there isn't one.
* Analogy: This is like a court convicting an innocent person.

* Symbol: The probability of committing a Type I error is denoted by α (alpha), which is also the significance level of your test (e.g., 0.05 or 5%).

* Example:
    
  **Scenario:** A pharmaceutical company is testing a new drug for a common cold. The null hypothesis (H0) is that the new drug has no effect on cold duration. The alternative hypothesis (Ha) is that the new drug reduces cold duration.
  
  **Type I Error:** The company concludes that the new drug is effective (rejects H0), but in reality, it has no effect. This could lead to a drug being marketed that is ineffective, potentially misleading patients and wasting resources.

**Type II Error (False Negative)**

* Definition: A Type II error occurs when you fail to reject a false Null Hypothesis. In simpler terms, you conclude there is no significant effect or difference when, in reality, there is one.

* Analogy: This is like a court letting a guilty person go free.

* Symbol: The probability of committing a Type II error is denoted by β (beta). The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.

* Example:

  **Scenario**: Using the same drug trial example, the null hypothesis (H0) is that the new drug has no effect on cold duration, and the alternative hypothesis (Ha) is that it reduces cold duration.
  **Type II Error:** The company concludes that the new drug is not effective (fails to reject H0), but in reality, it does reduce cold duration. This could mean a genuinely effective drug is discarded, and patients miss out on a beneficial treatment.


**Q3. What is power of statistical test?**

The power of a statistical test is a crucial concept in hypothesis testing. In simple terms, it's the probability that your test will correctly detect an effect when there is an effect to be detected.

**Definition**

  **Power (1 - β)**: The probability of correctly rejecting a false null hypothesis (H0).
  * It's the probability of
avoiding a Type II error (false negative).
  * A high power means that the test is sensitive enough to detect an effect when one truly exists.

**Understanding Power in Context**

Imagine you're conducting a study to see if a new drug improves a patient's condition. You set up your hypotheses:

* H0 (Null Hypothesis): The new drug has no effect on the patient's condition.
* Ha (Alternative Hypothesis): The new drug does have an effect on the patient's condition.

Now, consider the outcomes:

  1. **If the drug truly works (H0 is false)**, the power of your test is the probability that your study will correctly conclude that the drug works (i.e., you reject H0).
  2. **If the drug truly doesn't work (H0 is true)**, you want your test to correctly conclude that it doesn't work (i.e., you fail to reject H0). This is related to your significance level (alpha).

**Factors Affecting Power**

Several factors can influence the power of a statistical test:

1. Significance Level (α): Increasing α (e.g., from 0.01 to 0.05) makes it easier to reject the null hypothesis, thus increasing power. However, it also increases the risk of a Type I error.
2. Sample Size (n): Larger sample sizes lead to more precise estimates and smaller standard errors, which generally increases the power of the test.
3. Effect Size: This is the magnitude of the difference or relationship you're trying to detect. A larger effect size is easier to detect, leading to higher power.
4. Variability (Standard Deviation): Lower variability within the data makes it easier to detect an effect, thus increasing power. Higher variability makes it harder.
5. Type of Test: Some statistical tests are inherently more powerful than others under certain conditions.

In essence, power is a measure of the sensitivity of your statistical test to detect a real effect.

**Q4. Explain input and out put characteristics for  stats.ttest_rel function in your own words**

The scipy.stats.ttest_rel function is used to perform a paired t-test, which is designed to compare the means of two related samples. Think of it as comparing measurements taken from the same subjects or units under two different conditions (e.g., before and after an intervention).

**Input Characteristics**

The `ttest_rel` function typically takes two main arguments as input:

1. **`a` (First Sample/Measurements):**
    * This is usually an array-like
object (like a Python list or a NumPy array) containing the numerical observations from the first set of measurements.
    * Example: If you're measuring blood pressure before a new medication, a would be the list of blood pressure readings before the medication for all participants.

2. **`b` (Second Sample/Measurements):**
    * This is also an array-like object containing the numerical observations from the second set of measurements.
        Crucially, these measurements must be paired with the a measurements. This means a[i] and b[i] should correspond to the same subject or unit.
    * Example: Continuing the blood pressure example, b would be the list of blood pressure readings from the same participants after taking the medication.

**Key Requirements for Inputs:**

  * Paired Data: The most important characteristic is that the data in a and b must be paired or dependent. Each observation in a must have a corresponding, related observation in b.
 * Same Length: Consequently, a and b must have the exact same number of elements.
 * Numerical: Both inputs must consist of numerical data.

**Output Characteristics**

The ttest_rel function typically returns two values:

1. **`statistic` (The Calculated t-statistic):**
    * This is a single numerical value that quantifies the difference between the means of your two paired samples, relative to the variability within the differences.
    * In simpler terms: It tells you how big the observed difference is compared to the "noise" or spread in your data. A larger absolute value of the t-statistic generally indicates a greater difference between the means that is less likely due to random chance.

2. **`pvalue` (The Two-Sided p-value):**
      * This is a single numerical value, always between 0 and 1, that helps you determine the statistical significance of your results.
      * In simpler terms: The p-value tells you the probability of observing a difference as extreme as (or more extreme than) what you found, assuming there is actually no real difference between the paired samples in the population (i.e., assuming the null hypothesis is true).
      * How to interpret:
            
          * If the pvalue is very small (e.g., less than your chosen significance level, typically 0.05), you would usually conclude that the observed difference is statistically significant, and you would reject the null hypothesis. This suggests there is a real difference between the paired conditions.
          * If the pvalue is large (e.g., greater than 0.05), you would typically fail to reject the null hypothesis. This suggests that the observed difference could easily have occurred by random chance, and there isn't enough evidence to claim a real difference.

In essence, you feed ttest_rel two lists of related numbers, and it tells you if the average difference between those numbers is statistically meaningful.

**Q5. What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**

The `scipy.stats` module, while excellent for statistical tests and distributions, does not directly provide a single function for calculating the power of a statistical test.

However, in Python, the `statsmodels` library is the standard and most comprehensive tool for power and sample size calculations. Specifically, the `statsmodels.stats.power` module is designed for this purpose.

**`statsmodels.stats.power` Module**:

The statsmodels.stats.power module provides functions and classes to perform power analysis for various statistical tests (e.g., t-tests, ANOVA, chi-squared tests).

**Common Function for General Power Calculation: `solve_power`**

A very versatile function is statsmodels.stats.power.solve_power. This function can solve for one of four parameters (number of observations, effect size, significance level, or power) given the other three. This means you can use it to:

  * Calculate power given sample size, effect size, and alpha.
  * Calculate required sample size given power, effect size, and alpha.
  * Calculate detectable effect size given power, sample size, and alpha.
  * Calculate required alpha given power, sample size, and effect size.

It takes arguments like:

  * effect_size: The standardized effect size (e.g., Cohen's d for t-tests).
  * nobs1: Number of observations in one sample (for two-sample tests).
  * alpha: Significance level (Type I error rate).
  * power: The desired statistical power.
  * ratio: Ratio of sample sizes (nobs2/nobs1) for two-sample tests.
  * alternative: Specifies the alternative hypothesis ('two-sided', 'one-sided').
  * value: The parameter you want to solve for (e.g., 'power', 'nobs1', 'effect_size', 'alpha').

**Output Range and Significance when Solving for Power**

When you use solve_power (or similar power analysis functions/classes in statsmodels) to calculate the power, the output will be a single floating-point number representing the probability of correctly rejecting a false null hypothesis.

  * **Output Range:** The power value will always be between 0 and 1 (inclusive).
        Power = 0: Implies a 0% chance of detecting a true effect. The test is completely unable to detect the effect.
        Power = 1: Implies a 100% chance of detecting a true effect. The test is guaranteed to detect the effect (in practice, perfect power is rarely achieved).

  * **Significance:**
        Interpretation: A higher power value indicates a more sensitive test, meaning it's more likely to detect a real effect if one exists. Conversely, a lower power value means the test is less likely to detect a real effect, increasing the risk of a Type II error (false negative).
        Desirable Level: Researchers generally aim for a power of 0.80 (or 80%) as a conventional standard. This means there's an 80% chance of detecting a true effect if it exists, and a 20% chance of a Type II error (β = 0.20).
  * **Factors Influencing Power:** As discussed earlier, power is influenced by:
        
      1. Effect Size: Larger effects are easier to detect (higher power).
      2. Sample Size: Larger samples provide more information (higher power).
      3. Significance Level (α): A higher α (e.g., 0.10 vs. 0.05) increases power but also increases the risk of a Type I error.
      4. Variability: Lower variability in the data (smaller standard deviation) increases power.

while scipy.stats provides the statistical tests themselves, statsmodels.stats.power is the dedicated tool in Python for performing power analysis to understand the probability of your test detecting an effect of a certain magnitude, given your sample size and significance level.
