**## 2. Data Selection**

**Dataset Description**

The Iris dataset contains measurements of iris flowers from three different species:

* Setosa
* Versicolor
* Virginica

**Features (Attributes)**
| Attribute Name | Description          | Data Type   |
| -------------- | -------------------- | ----------- |
| sepal_length   | Length of sepal (cm) | Float       |
| sepal_width    | Width of sepal (cm)  | Float       |
| petal_length   | Length of petal (cm) | Float       |
| petal_width    | Width of petal (cm)  | Float       |
| species        | Type of iris flower  | Categorical |


**3. Import Libraries**


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

**4. Load Dataset**

```python
# Load iris dataset from seaborn
df = sns.load_dataset('iris')

# Display first 5 rows
df.head()
```

**5. Initial Data Inspection**

```python
# Shape of dataset
df.shape

# Data types
df.dtypes

# Summary statistics
df.describe()

# Check column names
df.columns
```

**6. Data Understanding**

* Total rows: 150
* Total columns: 5
* All feature columns are numerical except species
* No immediate inconsistencies visible

**7. Checking Missing Values**

```python
# Check missing values
df.isnull().sum()
```

**Observation:**

No missing values found in dataset

**Conclusion**:

Dataset is clean with respect to missing data


**9. Data Filtering (Selection of Attributes)**

Reasoning:

For this project, we keep all attributes because:

* Each feature contributes to classification
* No irrelevant columns present



```python
# Selecting all relevant columns
df_filtered = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']]
df_filtered.head()
```

**10. Data Type Validation**

```python
df_filtered.dtypes
```

**Observation:**

* Numerical features are correctly stored as float
* Species is categorical (object type)

**11. Data Transformation**

```python
df_filtered['species'] = df_filtered['species'].astype('category')
```

```python
df_filtered.dtypes
```

# 12. Feature Renaming

```python
df_filtered.rename(columns={
    'sepal_length': 'SepalLength',
    'sepal_width': 'SepalWidth',
    'petal_length': 'PetalLength',
    'petal_width': 'PetalWidth',
    'species': 'Species'
}, inplace=True)

df_filtered.head()
```

**13. Final Cleaned Dataset**

```python
df_filtered.info()
```

**Final Observations:**
* Dataset is clean
* No missing values
* No duplicates
* Proper data types assigned
* Ready for EDA

**14. Save Cleaned Data**

```python
df_filtered.to_csv('iris_cleaned.csv', index=False)
```

**Conclusion**:
* The Iris dataset was selected due to its simplicity and suitability for classification tasks
* Data cleaning steps ensured:
  * No missing values
  * No duplicates
