**Q1.What is need for Hypothesis Testing ?**
Hypothesis Testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

Need for Hypothesis Testing:

1. Decision Making:

It helps determine whether a claim or assumption about a dataset is true or false.

2. Testing Assumptions:

Researchers often make assumptions (hypotheses). Hypothesis testing checks whether these assumptions are statistically valid.

3. Data Validation:

It verifies whether the observed results are due to chance or a real effect.

4. Comparing Groups:

It helps compare different datasets or populations to see if there is a significant difference.

5. Reducing Uncertainty:

Hypothesis testing provides a systematic way to make conclusions with a known level of confidence.

6. Supporting Research and Analysis:

It is widely used in fields like data science, medicine, economics, and machine learning to validate results.

**Q2. With a neat diagram show important steps in hypothesis Testing.**

Important Steps in Hypothesis Testing:

1. Set Null Hypothesis (H₀):

The null hypothesis states that there is no effect, difference, or relationship in the population.

2. Set Alternative Hypothesis (H₁)
The alternative hypothesis represents the actual claim or research assumption, meaning there is an effect or difference.

3. Carry Out Significance Test
Select a significance level (α) such as 0.05 or 0.01 to determine the probability of rejecting the null hypothesis.

4. Obtain Test Statistic
Calculate the test statistic (e.g., z-value, t-value, or chi-square value) using the sample data.

5. Compare Test Statistic with Critical Value
Compare the calculated value with the critical value from statistical tables to determine whether it falls in the rejection region.

6. Obtain p-value
Determine the p-value, which indicates the probability of observing the sample result if the null hypothesis is true.

7. Make a Decision

If p-value ≤ α, reject the null hypothesis (H₀).

If p-value > α, fail to reject the null hypothesis.

**Q3.What is a student t distribution?**

The Student's t-distribution, often just called the t-distribution, is a type of probability distribution that is used to estimate population parameters when the sample size is small and the population variance is unknown. It is similar to the normal distribution but has fatter tails, meaning it has more probability in the tails than the normal distribution. This accounts for the increased uncertainty that comes with smaller sample sizes.

Key characteristics of the t-distribution:

1. Bell-shaped and Symmetric: Like the normal distribution, it is bell-shaped and symmetric around its mean of zero.
2. Degrees of Freedom (df): Its shape is determined by a parameter called 'degrees of freedom' (df), which is typically related to the sample size (e.g., df = n - 1, where n is the sample size).
3. Fatter Tails: As the degrees of freedom increase, the t-distribution approaches the normal distribution. With fewer degrees of freedom (smaller sample sizes), its tails are thicker, indicating a higher probability of extreme values.
4. Used in Hypothesis Testing: It is commonly used in hypothesis testing when comparing means of small samples (e.g., in a t-test) or when constructing confidence intervals for the population mean when the population standard deviation is unknown.

**Q4. What is level of significance ? What role t plays in hypothesis Testing**

The level of significance (α), also known as the alpha level, is a crucial concept in hypothesis testing. It represents the probability of rejecting the null hypothesis when it is actually true. In simpler terms, it's the risk you're willing to take of making a Type I error (a false positive).

Commonly used levels of significance are:

1. 0.05 (5%): This is the most common and indicates a 5% chance of incorrectly rejecting the null hypothesis.
2. 0.01 (1%): This is a more stringent level, indicating a 1% chance of a Type I error.
3. 0.10 (10%): A less stringent level, indicating a 10% chance of a Type I error.

The t-value (or t-statistic) plays a significant role in hypothesis testing, particularly when dealing with small sample sizes or when the population standard deviation is unknown. It is used in what is known as a t-test.

Here's its role:

1. Measuring the Difference: The t-value measures the difference between the observed sample mean and the hypothesized population mean (or the difference between two sample means) in units of standard error. A larger absolute t-value indicates a greater difference.
2. Determining Statistical Significance: The calculated t-value is compared to a critical t-value from the t-distribution table (or used to calculate a p-value). This comparison helps determine if the observed difference is statistically significant (i.e., unlikely to have occurred by chance if the null hypothesis were true) at a chosen level of significance (α).
3. Foundation for Decision Making: If the calculated t-value falls into the rejection region (defined by the critical t-value and α) or if its associated p-value is less than α, the null hypothesis is rejected. Otherwise, we fail to reject the null hypothesis.

In essence, the t-value helps quantify how strong the evidence is against the null hypothesis, guided by the characteristics of the t-distribution which accounts for uncertainty due to smaller sample sizes.

**Q5.Design a hypothesis testing for identifying average age of students in TE IT class is 19 or not.**

### Hypothesis Testing Design: Average Age of Students in TE IT Class

To test if the average age of students in a TE IT class is 19, we will follow the standard steps of hypothesis testing:

#### 1. Formulate the Null and Alternative Hypotheses

*   **Null Hypothesis ($H_0$)**: The average age of students in the TE IT class is 19 years.
    *   $H_0: \mu = 19$

*   **Alternative Hypothesis ($H_1$)**: The average age of students in the TE IT class is not 19 years.
    *   $H_1: \mu \neq 19$

This is a two-tailed test because we are interested if the average age is *not equal to* 19, meaning it could be either less than or greater than 19.

#### 2. Select the Significance Level (α)

The significance level (alpha, α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). A common choice is 0.05.

*   **Let's choose $\alpha = 0.05$ (5%)**.

This means we are willing to accept a 5% chance of incorrectly concluding that the average age is not 19, when it actually is 19.

#### 3. Choose the Appropriate Statistical Test

Since we are testing a hypothesis about a population mean ($\mu$) and typically the population standard deviation is unknown, the **t-test** for a single mean is appropriate. We will need to collect a sample of student ages from the TE IT class.

*   **Test Statistic**: t-statistic

#### 4. Collect Sample Data

To perform the test, we need to collect a random sample of student ages from the TE IT class. Let's say we collect ages from 'n' students.

*   **Data needed**: Sample size (n), sample mean ($\bar{x}$), and sample standard deviation (s).

#### 5. Calculate the Test Statistic

The t-statistic will be calculated using the following formula:

$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$

Where:
*   $\bar{x}$ = sample mean age
*   $\mu_0$ = hypothesized population mean (19)
*   s = sample standard deviation
*   n = sample size

The degrees of freedom for this test will be $df = n - 1$.

#### 6. Determine the Critical Value(s) or p-value

*   **Critical Value Approach**: For a two-tailed test with $\alpha = 0.05$ and $df = n-1$, we would find the critical t-values ($t_{critical}$) from the t-distribution table. The rejection regions would be $t < -t_{critical}$ or $t > t_{critical}$.

*   **p-value Approach**: Calculate the p-value associated with the calculated t-statistic. The p-value is the probability of observing a sample mean as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true.

#### 7. Make a Decision

*   **Using Critical Value**: If the calculated t-statistic falls into the rejection region (i.e., $t < -t_{critical}$ or $t > t_{critical}$), we **reject the null hypothesis**.

*   **Using p-value**: If $p-value \le \alpha$ (0.05), we **reject the null hypothesis**. If $p-value > \alpha$, we **fail to reject the null hypothesis**.

If we reject the null hypothesis, it means there is sufficient statistical evidence to conclude that the average age of students in the TE IT class is not 19. If we fail to reject the null hypothesis, it means there is not enough statistical evidence to conclude that the average age is different from 19.

**Q6.Check if you can accept Ho at alpha 95% ? Match your conclusions with python functions.**

To check whether we accept the Null Hypothesis (H₀) at 95% confidence level (α = 0.05), we compare the p-value with α.

Decision Rule

If p-value ≤ 0.05 → Reject H₀

If p-value > 0.05 → Accept (Fail to Reject) H₀

This rule is used in Hypothesis Testing.

Example using Python

In Python, we can verify the decision using statistical functions from SciPy.

Python Code Example (t-test):

```python
import numpy as np
from scipy import stats

# sample data
sample1 = [22, 25, 24, 23, 26]
sample2 = [20, 21, 19, 18, 20]

# perform t-test
t_stat, p_value = stats.ttest_ind(sample1, sample2)

alpha = 0.05

print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value > alpha:
    print("Accept (Fail to Reject) Null Hypothesis H0")
else:
    print("Reject Null Hypothesis H0")
```

Set α = 0.05 (95% confidence level).

Compute the p-value using Python statistical functions.

Compare p-value with α.

If p-value > 0.05, we accept H₀; otherwise, we reject H₀.

example using the Iris Dataset to check whether we accept the Null Hypothesis (H₀) at 95% confidence level (α = 0.05) using Python.

Hypothesis

H₀ (Null Hypothesis): The mean sepal length of two species is equal.

H₁ (Alternative Hypothesis): The mean sepal length of two species is different.

We perform a t-test using SciPy.

```python
import seaborn as sns
from scipy import stats

# load iris dataset
df = sns.load_dataset("iris")

# select two species
setosa = df[df["species"] == "setosa"]["sepal_length"]
versicolor = df[df["species"] == "versicolor"]["sepal_length"]

# perform t-test
t_stat, p_value = stats.ttest_ind(setosa, versicolor)

alpha = 0.05

print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value > alpha:
    print("Accept (Fail to Reject) Null Hypothesis H0")
else:
    print("Reject Null Hypothesis H0")
```

Sepal length values for setosa and versicolor species are selected.

A t-test is performed using SciPy.

The p-value is compared with α = 0.05.
