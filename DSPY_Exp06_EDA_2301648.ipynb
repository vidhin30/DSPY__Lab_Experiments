**Q2.In IRIS data set which non graphical EDA is performed ? **

For the Iris Dataset, Non-Graphical Exploratory Data Analysis (EDA) means analyzing the dataset without using plots or charts, only using statistical summaries and table-based analysis (usually done with Pandas).

**Non-Graphical EDA performed on the Iris dataset:**

1. Dataset Structure:

* Check number of rows and columns using shape.

* Iris dataset contains 150 tuples and 5 features.

2. Data Types of Attributes

* Use info() to identify the type of each column.

* Four features are numeric (sepal_length, sepal_width, petal_length, petal_width) and one is categorical (species).

3. Statistical Summary

* Use describe() to compute:

* Mean

* Median

* Standard Deviation

* Minimum and Maximum values

* Quartiles (25%, 50%, 75%)

4. Checking Missing Values

* Use isnull().sum() to identify missing data.

* The Iris dataset has no missing values.

5. Frequency Distribution

* Use value_counts() on the species column.

* Each species appears 50 times.

6. Correlation Analysis

* Use corr() to find relationships between numeric variables.

* Shows how sepal and petal measurements are related.


**Q3. In  Air Quality data set which graph is used to perform multivariate analysis?**

In the Air Quality Dataset, multivariate analysis is commonly performed using a Pair Plot.

* Graph Used:

   Pair Plot (Scatter Plot Matrix) – available in Seaborn.

Explanation

* A Pair Plot shows the relationship between multiple variables simultaneously.

* It creates a matrix of scatter plots for every pair of numerical variables in the dataset.

* The diagonal plots usually show the distribution (histogram or KDE) of each variable.

* This helps analyze correlations between air pollutants such as CO, NO₂, O₃, etc.

In the Air Quality dataset, Pair Plot (Scatter Plot Matrix) is used to perform multivariate analysis, as it visualizes relationships among multiple variables simultaneously.
